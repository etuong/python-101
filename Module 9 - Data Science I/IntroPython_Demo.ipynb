```python
df = pd.read_csv("sample_data/college_majors.csv")
# print the first 5 entries in the data file
df.head(5)

```

```python
df['Major']
```

```python
# list the names of the first 10 majors in alphabetical order
df.sort_values(by="Major")['Major'].head(10)
```

```python
# print the most popular major among men
popular_men = df[df.Men == df.Men.max()]
print(popular_men[["Major", "Men"]])
```

```python
# print the most popular major among women
popular_women = df[df.Women == df.Women.max()]
print(popular_women[["Major", "Women"]])
```

```python
top20_popular = df.sort_values(by="Total", ascending=False).head(20)
print(top20_popular[["Major", "Total"]])
```

```python
# 1. Create a new column that contains the percentage of people employed per major
# 2. Display the Major and EmployedPct of the top 20 majors
df["EmployedPct"] = df["Employed"]/df["Total"]
top20_employed = df.sort_values(by="EmployedPct", ascending=False).head(20)
print(top20_employed[["Major", "EmployedPct"]])
```

```python
# print the major(s) that is/are in both the 20 most popular and 20 most employed
# HINT: use the 'merge' function to perform an inner join on the 'Major' field between DataFrames
most_popular_employed = pd.merge(top20_popular, top20_employed, how="inner", on=["Major", "Major"])["Major"]
print(most_popular_employed)
```
